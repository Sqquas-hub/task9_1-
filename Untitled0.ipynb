```python
import pandas as pd
import matplotlib.pyplot as plt

# Не получалось загрузить файл с данными в гугл колаб для проверки
# Решил добавить команду для самостоятельного выбора файла на компьютере
from google.colab import files
uploaded = files.upload()

# Загрузка данных
df = pd.read_csv('books.csv', encoding='utf-8')

# Создание фигуры с двумя графиками
plt.figure(figsize=(12, 6))

# 1. Линейный график для 2023 года (количество книг)
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, 1-я позиция
df_2023 = df[df['Год'] == 2023]
plt.plot(df_2023['Месяц'], df_2023['Книги'],
         marker='o', color='red', label='Прочитано книг')
plt.grid(True)
plt.title('Прочитанные книги 2023')
plt.xlabel('Месяцы')
plt.ylabel('Количество книг')
plt.legend()

# 2. Столбчатая диаграмма для 2024 года (количество страниц)
plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, 2-я позиция
df_2024 = df[df['Год'] == 2024]
plt.bar(df_2024['Месяц'], df_2024['Страницы'], color='green')
plt.grid(True)
plt.title('Прочитанные страницы 2024')
plt.xlabel('Месяцы')
plt.ylabel('Количество страниц')

# Настройка отображения
plt.tight_layout()
plt.show()
```
